```python
%matplotlib inline
import numpy as np
from scipy.signal import ricker
import matplotlib.pyplot as plt
```

```python
#velocity model, a gradient velocity with depth
n_samples=600
n_traces=500
model1=np.zeros((n_samples,n_traces))
for j in range(0,n_samples):
    model1[j,:]=1500.+j*10.
model1[300:400,200:300]=1500.
plt.imshow(model1, cmap='Spectral')
plt.colorbar()
plt.title('Velocities')
plt.show()
```

```python
#calculating the TWT in the model, square grid with 1m (dz)
dz=1.
TWT=np.zeros((n_samples, n_traces));
TWT[0,:]=2.*dz/model1[0,:]
for j in range(1,n_samples):
    TWT[j,:]=TWT[j-1]+2.*dz/model1[j,:]
TMAX=max(TWT[-1,:])
plt.imshow(TWT, cmap='Spectral')
plt.colorbar()
plt.title('TWT')
plt.show()
```

```python
#resampling in a time rate multiple of 2ms
dt=2.e-4
TWT_rs=np.zeros(np.ceil(TMAX/dt))
for j in range(1,len(TWT_rs)):
    TWT_rs[j]=TWT_rs[j-1]+dt;
```

```python
rc=np.zeros((n_samples,n_traces));
rc[1:]=(model1[1:]-model1[:-1])/(model1[1:]+model1[:-1])
rc[0,:]=rc[1,:]
plt.imshow(rc, cmap='Spectral')
plt.colorbar()
plt.title('RC')
plt.show()
```

```python
plt.plot(TWT[:,250],rc[:,250],'.')
plt.title('RC - coluna 250')
plt.show()
```

```python
#linear interpolation of velocity
from scipy import interpolate

rc_tsl=np.ones((np.ceil(TMAX/dt), n_traces))

j=250
lim=np.ceil(TWT[-1,j]/dt)-1;
tt=rc[:,j]

tck=interpolate.interp1d(TWT[:,j],tt)
rc_tsl[10:lim,j]=tck(TWT_rs[10:lim])

rc_tsl[lim:,j]=rc_tsl[lim-1,j]
rc_tsl[0:10,j]=rc_tsl[10,j]
```

```python
plt.plot(TWT[:,j],rc[:,j],'k*',TWT_rs,rc_tsl[:,j],'.')
plt.legend(['Original','Interpolated'],loc='best')
plt.title('RC - linear intepolation')
plt.show()
```

```python
#spline interpolation of velocity
from scipy import interpolate

rc_tsspl=np.ones((np.ceil(TMAX/dt), n_traces))
j=250
lim=np.ceil(TWT[-1,j]/dt)-1;

tt=rc[:,j]

tck= interpolate.splrep(TWT[:,j],tt,s=0)
rc_tsspl[0:lim,j]=interpolate.splev(TWT_rs[0:lim],tck,der=0)

rc_tsspl[lim:,j]=rc_tsspl[lim-1,j]
rc_tsspl[0:10,j]=rc_tsspl[10,j]

plt.plot(TWT[:,j],rc[:,j],'k*',TWT_rs,rc_tsspl[:,j],'.')
plt.legend(['Original','Interpolated'],loc='best')
plt.title('RC - spline intepolation')
plt.show()
```

```python
plt.plot(rc_tsspl[:,j],rc_tsl[:,j],'.')
plt.title('Comparing spline x linear')
plt.show()
```

```python
#resampling to 2ms
rc_spl=np.zeros((np.ceil(TMAX/0.002)))
rc_l=np.zeros((np.ceil(TMAX/0.002)))
TWT_ts=np.zeros((np.ceil(TMAX/0.002)))

rc_spl[0]=rc_tsspl[0,250]
rc_l[0]=rc_tsl[0,250]
for j in range(1,len(TWT_ts)):
    rc_spl[j]=rc_tsspl[10*j,250]  #10=dt/dt_new, dt_new=0.002=2ms
    rc_l[j]=rc_tsl[10*j,250] 
    TWT_ts[j]=TWT_rs[10*j] 

#plt.plot(TWT_rs,model1_tsl[:,100],'.')
plt.plot(TWT_ts,rc_l,'.',TWT_ts,rc_spl,'*r')
plt.title('RC resampled to 2ms')
plt.show()
```

```python
f=20;dt=2.e-3;

nw=2.2/f/dt;
nw=2*np.floor(nw/2)+1;

nc=np.floor(nw/2);
w = np.zeros(nw);
k=np.arange(1,nw+1);

alpha = (nc-k+1)*f*dt*np.pi;
beta=alpha**2;
w = (1.-beta*2)*np.exp(-beta);

plt.plot(w)
plt.show()
```

```python
synth_l=np.convolve(rc_l,w,mode='same')
plt.plot(TWT_ts,synth_l,'-',TWT_ts,rc_l,'.')
plt.title('Synthethic - Interp linear velocity')
plt.show()
```

```python
synth_spl=np.convolve(rc_spl,w,mode='same')
plt.plot(TWT_ts,synth_spl,'-',TWT_ts,rc_spl,'.')
plt.title('Synthethic - Interp spline velocity')
plt.show()
```
