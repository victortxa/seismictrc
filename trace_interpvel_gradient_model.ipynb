```python
#calculating the TWT in the model, square grid with 1m (dz)
dz=1.
TWT=np.zeros((n_samples, n_traces));
TWT[0,:]=2.*dz/model1[0,:]
for j in range(1,n_samples):
    TWT[j,:]=TWT[j-1]+2.*dz/model1[j,:]
TMAX=max(TWT[-1,:])
plt.imshow(TWT, cmap='Spectral')
plt.colorbar()
plt.title('TWT')
plt.show()
```

```python
#resampling in a time rate multiple of 2ms
dt=2.e-4
TWT_rs=np.zeros(np.ceil(TMAX/dt))
for j in range(1,len(TWT_rs)):
    TWT_rs[j]=TWT_rs[j-1]+dt;
```

```python
#linear interpolation of velocity
from scipy import interpolate

model1_tsl=np.ones((np.ceil(TMAX/dt), n_traces))

j=250
lim=np.ceil(TWT[-1,j]/dt)-1;
tt=model1[:,j]

tck=interpolate.interp1d(TWT[:,j],tt)
model1_tsl[10:lim,j]=tck(TWT_rs[10:lim])

model1_tsl[lim:,j]=model1_tsl[lim-1,j]
```

```python
plt.plot(TWT[:,j],model1[:,j],'k-',TWT_rs,model1_tsl[:,j],'.')
plt.legend(['Original','Interpolated'],loc='best')
plt.title('Velocity Model - linear intepolation')
plt.show()
```

```python
#spline interpolation of velocity
from scipy import interpolate

model1_tsspl=np.ones((np.ceil(TMAX/dt), n_traces))
j=250
lim=np.ceil(TWT[-1,j]/dt)-1;

tt=model1[:,j]

tck= interpolate.splrep(TWT[:,j],tt,s=0)
model1_tsspl[0:lim,j]=interpolate.splev(TWT_rs[0:lim],tck,der=0)

model1_tsspl[lim:,j]=model1_tsspl[lim-1,j]

plt.plot(TWT[:,j],model1[:,j],'-',TWT_rs,model1_tsspl[:,j],'.')
plt.legend(['Original','Interpolated'],loc='best')
plt.title('Velocity Model - spline intepolation')
plt.show()
```

```python
#resampling to 2ms
vel_tsspl=np.zeros((np.ceil(TMAX/0.002)))
vel_tsl=np.zeros((np.ceil(TMAX/0.002)))
TWT_ts=np.zeros((np.ceil(TMAX/0.002)))

vel_tsspl[0]=model1_tsspl[0,100]
vel_tsl[0]=model1_tsl[0,100]
for j in range(1,len(TWT_ts)):
    vel_tsspl[j]=model1_tsspl[10*j,250]  #10=dt/dt_new, dt_new=0.002=2ms
    vel_tsl[j]=model1_tsl[10*j,250] 
    TWT_ts[j]=TWT_rs[10*j] 

#ynth_rs[:,100]=np.convolve(synth_rs[:,100],[0.25,0.5,0.25],mode='same')

#plt.plot(TWT_rs,model1_tsl[:,100],'.')
plt.plot(TWT_ts,vel_tsl,'.',TWT_ts,vel_tsspl,'-r')
plt.title('Velocity resampled to 2ms')
plt.show()
```
